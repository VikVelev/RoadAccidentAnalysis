```python
import pandas as pd
```

```python
data = pd.read_csv('dataset.csv')
```

```python
data.head()
```

```python
relevant_cols = ['Light_Conditions', 'Day_of_Week', 'Urban_or_Rural_Area', 'Road_Surface_Conditions', 'Weather_Conditions']
casualties = ['Number_of_Casualties']
```

```python
multivar_coefficients = []

corr_mat = data[relevant_cols + casualties].corr()
print(corr_mat[casualties[0]])
for h in corr_mat[casualties[0]]:
    print(h)

multivar_coefficients += list(
    map(
        lambda c: (list(c[0]), c[1]),
        zip(
            map(
                lambda c: (c, casualties[0]), 
                corr_mat.columns), 
        corr_mat[casualties[0]])))
```

```python
from itertools import combinations

perms = []
for i in range(2, len(relevant_cols)):
    perm = combinations(relevant_cols, i)
    perms += list(map(lambda p: list(p), perm))
```

```python
import numpy as np

def multivar_coeff(corr_mat):
    upper = np.triu(corr_mat, k=1)
    print(upper)
    return np.sqrt(
        np.sum(upper**2) - 2*np.product(upper))
```

```python

for perm in perms:
    
    variables = perm + casualties
    corr_mat = data[variables].corr().to_numpy()
    
    multivar_coefficients.append((variables, multivar_coeff(corr_mat)))

multivar_coefficients
```
